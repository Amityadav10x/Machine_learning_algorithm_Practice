**Importent Libreries**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix
```

```python
df = pd.DataFrame(data)
```

**Encoding of target feature**

```python
encoder = LabelEncoder()
df['Ethnicity'] = encoder.fit_transform(df['Ethnicity'])
```

**Splitted data feature and target**

```python
X = df.drop('Will survive till 70', axis=1)
y = df['Will survive till 70']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**Training and testing of data**

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

**SVM Classifier**

```python
svm_classifier = SVC(kernel='linear', random_state=42)
```

**Trained SVM Classifier**

```python
svm_classifier.fit(X_train, y_train)
```

```python
y_pred = svm_classifier.predict(X_test)
```

**Accuracy result of y_test**

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

**Classification Report of X_test**

```python
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
classes = ['No', 'Yes']
precision = [0.00, 0.50]
recall = [0.00, 1.00]
f1_score = [0.00, 0.67]

# Ploting the metrics using a bar chart

x = np.arange(len(classes))  # Created an array for x-axis positions
width = 0.20

fig, ax = plt.subplots(figsize=(8, 6))
rects1 = ax.bar(x - width, precision, width, label='Precision')
rects2 = ax.bar(x, recall, width, label='Recall')
rects3 = ax.bar(x + width, f1_score, width, label='F1-Score')

# Added labels, title, and legend
ax.set_xlabel('Classes')
ax.set_ylabel('Scores')
ax.set_title('Classification Report Metrics')
ax.set_xticks(x)
ax.set_xticklabels(classes)
ax.legend()

# Showing the plot
plt.xticks(rotation=40, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
